Las peliculas que ha alquilado cada cliente

```python
cursor = conexion.cursor()
# Ejecutar la consulta
cursor.execute("""SELECT c.first_name, c.last_name, f.title
                FROM customer c
                JOIN rental r ON c.customer_id = r.customer_id
                JOIN inventory i ON r.inventory_id = i.inventory_id
                JOIN film f ON i.film_id = f.film_id""")
                
# Obtener los resultados
resultado = cursor.fetchall()
# Cerrar la conexión
conexion.close()
# Mostrar los resultados
for fila in resultado:
    print(fila)
```

Contar cuantas peliculas ha alquilado cada cliente

```python
cursor = conexion.cursor()
# Ejecutar la consulta
cursor.execute("""SELECT c.first_name, c.last_name, COUNT(r.rental_id) AS rental_count
                FROM customer c
                JOIN rental r ON c.customer_id = r.customer_id
                GROUP BY c.customer_id""")
                
# Obtener los resultados
resultado = cursor.fetchall()
# Cerrar la conexión
conexion.close()
# Mostrar los resultados
for fila in resultado:
    print(fila)
```

Mostrar solo los clientes que han alquilado mas de 10 peliculas


```python
cursor = conexion.cursor()
# Ejecutar la consulta
cursor.execute("""SELECT c.first_name, c.last_name, COUNT(r.rental_id) AS Total_Rentas
                FROM customer c
                JOIN rental r ON c.customer_id = r.customer_id
                GROUP BY c.customer_id
                HAVING Total_Rentas > 10""")
                
# Obtener los resultados
resultado = cursor.fetchall()
# Cerrar la conexión
conexion.close()
# Mostrar los resultados
for fila in resultado:
    print(fila)
```

cliente que ha alquilado la película con el film_id más alto.

```python
cursor = conexion.cursor()
# Ejecutar la consulta
cursor.execute("""SELECT c.first_name, c.last_name, c.email
                FROM customer c
                WHERE c.customer_id = (
                                    SELECT r.customer_id
                                    FROM rental r
                                    JOIN inventory i ON r.inventory_id = i.inventory_id
                                    WHERE i.film_id = (
                                                    SELECT MAX(film_id)
                                                    FROM film
                                            )
                                            LIMIT 1
                                        )""")
                
# Obtener los resultados
resultado = cursor.fetchall()
# Cerrar la conexión
conexion.close()
# Mostrar los resultados
for fila in resultado:
    print(fila)
```

Mostrar todas las categorias de peliculas sin repetir

```python
cursor = conexion.cursor()
# Ejecutar la consulta
cursor.execute("""SELECT DISTINCT c.name
                FROM category c
                JOIN film_category fc ON c.category_id = fc.category_id
                JOIN film f ON fc.film_id = f.film_id""")
                
# Obtener los resultados
resultado = cursor.fetchall()
# Cerrar la conexión
conexion.close()
# Mostrar los resultados
for fila in resultado:
    print(fila)
```

Ver el titulo de cada pelicula y cuantas veces ha sido alquilada

```python
cursor = conexion.cursor()
# Ejecutar la consulta
cursor.execute("""SELECT f.title, COUNT(f.film_id) AS Contador_peliculas
                FROM film f
                JOIN inventory i ON i.film_id = f.film_id
                JOIN rental r ON i.inventory_id = r.inventory_id
                GROUP BY f.film_id
                ORDER BY Contador_peliculas DESC""")
                
# Obtener los resultados
resultado = cursor.fetchall()
# Cerrar la conexión
conexion.close()
# Mostrar los resultados
for fila in resultado:
    print(fila)
```
